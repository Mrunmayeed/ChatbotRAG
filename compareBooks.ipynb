```python
import pandas as pd
import numpy as np

from numpy.linalg import norm
```

```python
df = pd.read_csv('goodreads_data.csv').drop(['Unnamed: 0'],axis=1)
```

```python
df = df[['Book','Genres','Description']]
df
```

```python
all_genres_ = []

for i,v in df['Genres'].iteritems():
  arr = v[2:-2].split("', '")
  all_genres_.extend(arr)


all_genres = list(set(all_genres_))
len_gen = len(all_genres)
genres = dict(zip(all_genres,np.arange(len_gen)))
```

```python
from collections import Counter


all_genres_ = Counter(all_genres_)
all_genres_ = dict(sorted(all_genres_.items(), key=lambda item: item[1], reverse=True))
```

```python
df['Description Length'] = df.Description.str.len()
df['Number of Genres'] = df.Genres.str.split(", ").str.len()
df.describe()
```

```python
def get_genre_vector(glist):
  feature = np.zeros(len_gen)
  glist = glist[2:-2].split("', '")
  for g in glist:
    feature[genres[g]] = 1
  return feature

df['genre_vector'] = df['Genres'].apply(get_genre_vector)
```
